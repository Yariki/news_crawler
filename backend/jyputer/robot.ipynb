```python
from protego import Protego
import urllib

robots_url = "https://www.bbc.com/robots.txt"

response = urllib.request.urlopen(robots_url)
robots_content = response.read().decode('utf-8')

robot = Protego.parse(robots_content)
print(robot.can_fetch("*", "https://www.bbc.com/"))
print(robot.can_fetch("*", "https://www.bbc.com/news/articles/c202zn5gg0lo"))
print(robot.can_fetch("*", "https://www.bbc.com/public/"))
print(robot.crawl_delay("*"))
print(robot.request_rate("*"))
for sitemap in robot.sitemaps:
    print(sitemap)
```

```python
import urllib.robotparser


robots_url = "https://www.bbc.com/robots.txt"

response = urllib.request.urlopen(robots_url)
robots_content = response.read().decode('utf-8')
print(robots_content)

robot = urllib.robotparser.RobotFileParser()
robot.set_url(robots_url)
robot.read()
print(robot.can_fetch("*", "https://www.bbc.com/"))
print(robot.can_fetch("*", "https://www.bbc.com/private/"))
print(robot.can_fetch("*", "https://www.bbc.com/public/"))
print(robot.crawl_delay("*"))
print(robot.request_rate("*"))
print(robot.site_maps())

print(robot.read())





```
